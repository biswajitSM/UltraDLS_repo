```python
def comp_FCS(file1, file2, tmin=0, tmax=10):
    fig, ax = subplots(figsize=(10, 6))
    
    df_fcs = pd.read_csv(file1, index_col=False, names=None, skiprows=1, header=None, sep='\s+', engine='python');
    df_fcs = df_fcs[df_fcs[0]>=tmin];
    df_fcs = df_fcs[df_fcs[0]<=tmax];
    df_fcs
    ax.plot(df_fcs[0], df_fcs[1], label=file1)
    
    df_fcs = pd.read_csv(file2, index_col=False, names=None, skiprows=1, header=None, sep='\s+', engine='python');
    df_fcs = df_fcs[df_fcs[0]>=tmin];
    df_fcs = df_fcs[df_fcs[0]<=tmax];
    df_fcs
    ax.plot(df_fcs[0], df_fcs[1], label=file2)
    ax.set_xscale('log')
    ax.set_xlabel('Time[ms]')
    ax.set_ylabel(r'$G(\tau)-1$')
    ax.set_xlim(0.0004, tmax)
    ax.legend()
```

```python
def comparespec(AuNR_a, AuNR_b, AuNR_c, skiplines=300):
    fig, ax = plt.subplots()
    
    #first nanorod (a)
    df_a = pd.read_csv(AuNR_a, skiprows=skiplines, header=None, sep = '\t', engine='python');
    df_a = df_a[df_a[1]<7*average(df_a[1])]
    L_max=df_a[1].idxmax();L_max=df_a[0][L_max];
    ax.plot(df_a[0], df_a[1]/max(df_a[1]),'-', label=AuNR_a+'_SPR_'+str(round(L_max, 0))+' nm')

    #first nanorod (b)
    df_b = pd.read_csv(AuNR_b, skiprows=skiplines, header=None, sep = '\t', engine='python');
    df_a = df_b
    df_a = df_a[df_a[1]<7*average(df_a[1])]
    L_max=df_a[1].idxmax();L_max=df_a[0][L_max];
    ax.plot(df_a[0], df_a[1]/max(df_a[1]),'-', label=AuNR_b+'_SPR_'+str(round(L_max, 0))+' nm')
    
    df_b = pd.read_csv(AuNR_c, skiprows=skiplines, header=None, sep = '\t', engine='python');
    df_a = df_b
    df_a = df_a[df_a[1]<7*average(df_a[1])]
    L_max=df_a[1].idxmax();L_max=df_a[0][L_max];
    ax.plot(df_a[0], df_a[1]/max(df_a[1]),'-', label=AuNR_c+'_SPR_'+str(round(L_max, 0))+' nm')
    
    xlim(550, 800)
    legend(fontsize=8, bbox_to_anchor=(1.1,-0.1))
    grid(which='both', axis='x')
    # os.chdir(parentdir)   
os.chdir(parentdir)
```
